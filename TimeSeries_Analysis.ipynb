## Q4)Plot the monthly and yearly trends.

```python
data["total_emp"] = data.sum(axis=1)/(len(list(data))-1)
data_month = data.resample("M").mean()
fig, ax = plt.subplots(figsize=(15,6))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
ax.bar(data_month['2000':].index, data_month.loc['2000':, "total_emp"], width=25, align='center')
plt.xlabel("Months",fontsize = 12)
plt.ylabel("Unemployment rate",fontsize = 12)
plt.grid()
plt.show()
```

## Q5)Apply time series decomposition to your dataset to visualize the trend and seasonality

```python
from pylab import rcParams
import statsmodels.api as sm
rcParams['figure.figsize'] = 11, 7
decomposition = sm.tsa.seasonal_decompose(data_month['total_emp'], model='Additive')

plt.subplot(2,1,1)
fig = decomposition.trend.plot()
plt.title('Trend',fontsize = 13)
plt.xlabel("Years",fontsize = 12)

plt.subplot(2,1,2)
fig = decomposition.seasonal.plot()
plt.title('Seasonal',fontsize = 13)
plt.xlabel("Years",fontsize = 12)

plt.tight_layout()
plt.show()
```

## Q6)Visualize the seasonality ofAgriculture, Health and Finance sector.

```python
rcParams['figure.figsize'] = 9, 10

plt.subplot(3,1,1)
decomposition = sm.tsa.seasonal_decompose(data_month['Finance'], model='Additive')
fig = decomposition.seasonal.plot(color = 'brown')
plt.title('Finance',fontsize = 13)
plt.xlabel("Years",fontsize = 12)

plt.subplot(3,1,2)
decomposition = sm.tsa.seasonal_decompose(data_month['Agriculture'], model='Additive')
fig = decomposition.seasonal.plot(color = 'black')
plt.title('Agriculture',fontsize = 13)
plt.xlabel("Years",fontsize = 12)

plt.subplot(3,1,3)
decomposition = sm.tsa.seasonal_decompose(data_month['Education and Health'], model='Additive')
fig = decomposition.seasonal.plot(color = 'coral')
plt.title('Education and Health',fontsize = 13)
plt.xlabel("Years",fontsize = 12)

plt.tight_layout()
plt.show()
```

## Q7)Visualize the seasonality of multiple time series and the correlation between each time series in the dataset.

```python
data_quater = data.resample("Q").mean()
data_year = data.resample("Y").mean()
```

```python
fig, ax = plt.subplots(figsize=(20, 6))
ax.plot(data.loc[:, 'total_emp'], marker='.', linestyle='-', linewidth = 0.5, label='Monthly', color='black')
ax.plot(data_quater.loc[:, 'total_emp'], marker='o', markersize=8, linestyle='-', label='Quaterly', color='coral')
ax.plot(data_year.loc[:, 'total_emp'], marker='o', markersize=8, linestyle='-', label='Yearly', color='green')

ax.set_ylabel("Open")
ax.legend()
plt.show()
```
